#Changing Contrast and Brightness

```python
import numpy as np
```

```python
new_image = np.zeros(img1.shape, img1.dtype)
```

```python
contrast = 3.0
```

```python
bright = 2
```

```python
for y in range(img1.shape[0]):
    for x in range(img1.shape[1]):
      for c in range(img1.shape[2]):
        new_image[y,x,c] = np.clip(contrast*img1[y,x,c] + bright, 0, 255)
```

# Dilation

```python
#Define erosion size
s1 = 0
s2 = 10
s3 = 10
```

```python
#Define erosion type
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
```

```python
#Define and save erosion template
tmp1 = cv2.getStructuringElement(t1, (2*s1 + 1, 2*s1 + 1), (s1, s1))
tmp2 = cv2.getStructuringElement(t2, (2*s2 + 1, 2*s2 + 1), (s2, s2))
tmp3 = cv2.getStructuringElement(t3, (2*s3 + 1, 2*s3 + 1), (s3, s3))
```

```python
#Apply the erosion template to the image and save in different variables
final1 = cv2.erode(img1, tmp1)
final2 = cv2.erode(img1, tmp2)
final3 = cv2.erode(img1, tmp3)
```

```python
io.imshow(final1)
```

```python
io.imshow(final2)
```

```python
io.imshow(final3)
```

#EROSION CODE

```python
#define dilation size
d1 = 0
d2 = 10
d3 = 20
```

```python
#define dilation type
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
```

```python
#store the dilation template
tmp1 = cv2.getStructuringElement(t1, (2*d1 + 1, 2 * d1 +1), (d1, d1))
tmp2 = cv2.getStructuringElement(t2, (2*d2 + 1, 2 * d2 +1), (d2, d2))
tmp3 = cv2.getStructuringElement(t3, (2*d3 + 1, 2 * d3 +1), (d3, d3))
```

```python
#Apply dilation to the images
final1 = cv2.dilate(img1, tmp1)
final2 = cv2.dilate(img1, tmp2)
final3 = cv2.dilate(img1, tmp3)
```

```python
io.imshow(final1)
```

```python
io.imshow(final2)
```

```python
io.imshow(final3)
```

# Effecting Image Thresholding

```python
img1 = cv2.imread('cat1.jpg') # Reload the original image
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) # Convert to grayscale for thresholding

_, bin_normal    = cv2.threshold(img1_gray, 50, 225, cv2.THRESH_BINARY)
_, bin_inv       = cv2.threshold(img1_gray, 50, 225, cv2.THRESH_BINARY_INV)
_, truncated     = cv2.threshold(img1_gray, 50, 225, cv2.THRESH_TRUNC)
_, to_zero       = cv2.threshold(img1_gray, 50, 225, cv2.THRESH_TOZERO)
_, to_zero_inv   = cv2.threshold(img1_gray, 50, 225, cv2.THRESH_TOZERO_INV)
```
